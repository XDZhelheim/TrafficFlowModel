````python
import pandas as pd
import os

CACHE_PATH="../../Bigscity-LibCity/libcity/cache/"

%pwd
````

````python
# sz_taxi_5day

run_id_list=os.listdir(CACHE_PATH)[1:]

df_eval=pd.DataFrame(columns=["MAE", "MAPE", "MSE", "RMSE", "Masked MAE", "Masked MAPE", "Masked MSE", "Masked RMSE", "R2", "EVAR"])

for run_id in run_id_list:
    for file in os.listdir(os.path.join(CACHE_PATH, str(run_id), "evaluate_cache")):
        if file.endswith(".csv"):
            csv_res=file
    model_name=csv_res.split("_")[-3]
    df_eval_res=pd.read_csv(os.path.join(CACHE_PATH, str(run_id), "evaluate_cache", csv_res))
    
    df_eval.loc[model_name]=df_eval_res.mean().values

df_eval
````

````python
df_eval=pd.DataFrame(columns=["MAE", "MAPE", "MSE", "RMSE", "Masked MAE", "Masked MAPE", "Masked MSE", "Masked RMSE", "R2", "EVAR"])

os.chdir("/home/cseadmin/dz/Bigscity-LibCity")
for log_file in os.listdir("log_sz_taxi_202006"):
    model=log_file.split("_")[1].split(".")[0]
    with open(os.path.join("log_sz_taxi_202006", log_file)) as f:
        for line in f.readlines():
            if "INFO" not in line:
                continue
            if line.split("INFO")[1].startswith(" - Evaluate result"):
                res_path=line.split(" ")[-1].strip()
                break
    df_eval_res=pd.read_csv(res_path)
    
    df_eval.loc[model]=df_eval_res.mean().values
    
df_eval.sort_values("Masked MAPE")
````
