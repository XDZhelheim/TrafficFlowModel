```python
import pandas as pd
import os

CACHE_PATH="../../Bigscity-LibCity/libcity/cache/"

run_id_list=[79491, 44143, 63935, 12831, 41845, 72849, 23406, 9906, 17045, 91041, 12570, 29338, 13447]

%pwd
```

```python
df_eval=pd.DataFrame(columns=["MAE", "MAPE", "MSE", "RMSE", "Masked MAE", "Masked MAPE", "Masked MSE", "Masked RMSE", "R2", "EVAR"])

for run_id in run_id_list:
    for file in os.listdir(os.path.join(CACHE_PATH, str(run_id), "evaluate_cache")):
        if file.endswith(".csv"):
            csv_res=file
    model_name=csv_res.split("_")[-3]
    df_eval_res=pd.read_csv(os.path.join(CACHE_PATH, str(run_id), "evaluate_cache", csv_res))
    
    df_eval.loc[model_name]=df_eval_res.mean().values

df_eval
```
